### HyperParamter Tuning and Cross Validation

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty, C=c, solver=solver)
```

```python
params
```

### Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_test, y_pred)
print(score)

cm=confusion_matrix(y_test, y_pred)
print(cm)

print(classification_report(y_test, y_pred))
```

### Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomCV=RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')
```

```python
randomCV.fit(X_train, y_train)
```

```python
randomCV.best_score_
```

```python
randomCV.best_params_
```

```python
y_pred=randomCV.predict(X_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_test, y_pred)
print(score)

cm=confusion_matrix(y_test, y_pred)
print(cm)

print(classification_report(y_test, y_pred))
```

## Logistic Regression for Imbalanced data

```python
from collections import Counter
```

```python
X,y=make_classification(n_samples=10000, n_features=2, n_clusters_per_class=1, n_redundant=0, weights=[0.99], random_state=10)
```

```python
X
```

```python
y
```

```python
Counter(y)
```

```python
import seaborn as sns
import pandas as pd
```

```python
sns.scatterplot(pd.DataFrame(X)[0], pd.DataFrame(X)[1], hue=y)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# Hyperparameters tuning

model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
class_weight=[{0:w, 1:y} for w in [1,10, 50, 100] for y in [1, 10, 50, 100]]
```

```python
params=dict(penalty=penalty, C=c, solver=solver, class_weight=class_weight)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_test, y_pred)
print(score)

cm=confusion_matrix(y_test, y_pred)
print(cm)

print(classification_report(y_test, y_pred))
```

## Logistic Regression with ROC Curve and ROC AUC Score 

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
X,y=make_classification(n_samples=10000, n_classes=2, random_state=10)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# Created a dummy model with default 0 as output
dummy_model_prob=[0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
LR=LogisticRegression()
```

```python
LR.fit(X_train, y_train)
```

```python
model_prob=LR.predict_proba(X_test)
```

```python
model_prob=model_prob[:,1]
```

```python
model_prob
```

```python
dummy_model_auc=roc_auc_score(y_test, dummy_model_prob)

model_auc=roc_auc_score(y_test, model_prob)

print(dummy_model_auc)
```

```python
print(model_auc)
```

```python
dummy_fpr, dummy_tpr,_=roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, thresholds=roc_curve(y_test, model_prob)
```

```python
plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic')
plt.xlabel('False Postive Rate')
plt.xlabel('True Postive Rate')

plt.legend()
plt.show()
```
